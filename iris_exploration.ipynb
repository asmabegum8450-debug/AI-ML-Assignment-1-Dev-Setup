```python
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris(as_frame=True)   
df = iris.frame                  
df.head()                         
```

```python
df.info()
```

```python
df.describe()
df['target'].value_counts(), iris.target_names
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df['petal length (cm)'], bins=20)
plt.title("Histogram â€” Petal Length (cm)")
plt.xlabel("Petal Length (cm)"); plt.ylabel("Count")
plt.show()
```

```python
df_named = df.copy()
df_named['species'] = df_named['target'].map(dict(enumerate(iris.target_names)))

sns.boxplot(data=df_named, x='species', y='sepal width (cm)')
plt.title("Sepal Width by Species")
plt.xlabel("Species"); plt.ylabel("Sepal Width (cm)")
plt.show()
```

```python
sns.scatterplot(
    data=df_named, x='petal length (cm)', y='petal width (cm)',
    hue='species'
)
plt.title("Petal Length vs Petal Width â€” Colored by Species")
plt.xlabel("Petal Length (cm)"); plt.ylabel("Petal Width (cm)")
plt.show()
```
